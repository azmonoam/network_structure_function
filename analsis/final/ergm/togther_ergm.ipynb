```python
fig, (ax) = plt.subplots(4, 3, figsize=(14, 12), sharex='col')
c = [[blues[1], blues[2]], [rads[2], rads[3]], [COLORS[3], greens[0]]]
nums = [['A', 'B', 'C', 'D'], ['E', 'F', 'G', 'H'], ['I', 'J', 'K', 'L',], ]
for j, task in zip(range(ax.shape[1]), res.keys()):
    for i, vals in zip(range(ax.shape[0]), res[task].values()):
        n1, bins, patches = ax[i,j].hist(vals, color=c[j][0], density=True, bins=30, )
        ax[i,j].vlines(vals.mean(), ymin=0, ymax=max(n1), colors=c[j][1])
        print(vals.std())
        if round(vals.std(),2) < 0.01:
            textstr = '\n'.join((
                r'$\mu=%.2f$' % (vals.mean(),),
                r'$\sigma<0.01$'))
        else:
            textstr = '\n'.join((
                r'$\mu=%.2f$' % (vals.mean(),),
                r'$\sigma=%.2f$' % (vals.std(),)))
        ax[i,j].text(0.05, 0.95, textstr, transform=ax[i,j].transAxes, fontsize=11,                   verticalalignment='top', )
        if i == 3:
            ax[i,j].set_xlabel('Mean performance', fontsize=14)
        if j==0:
            ax[i,j].set_ylabel('Relative number\nof networks', fontsize=14)
        ax[i, j].tick_params(axis='both', labelsize=12)
        #ax[i, j].text(1.04, 1.04, nums[j][i], transform=ax[i, j].transAxes, fontsize=14, verticalalignment='top',
         #              weight="bold")
        fig.align_ylabels(ax[i:])
plt.savefig(
    f"{plot_path}/{time_str}_all_ergm_vs_other_distd.png",  bbox_inches='tight')
plt.show()
```
