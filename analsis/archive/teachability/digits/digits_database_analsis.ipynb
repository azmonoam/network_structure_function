```python
from datetime import datetime as dt

import pandas as pd

from analsis.analsis_utils.plot_utils import plot_binned_metric_vs_mean_metric, plot_two_metrics

import matplotlib.pyplot as plt
```

```python
results_path = "/Volumes/noamaz/modularity/teach_archs/digits"
local_base_path = '/Users/noamazmon/PycharmProjects/network_modularity'
results_csv_name = "2023-06-26-12-59-26_all_results_from_digits_teach_archs_results_no_duplicates.csv"
results = pd.read_csv(f"{results_path}/{results_csv_name}")
mean_metric_name = "mean_performance"
task = "digits"
time_str = dt.now().strftime("%Y-%m-%d-%H-%M-%S")
plot_path = f'{local_base_path}/plots/{task}_teachbility_plots'
```

```python
plot_binned_metric_vs_mean_metric(
    metric_name="connectivity_ratio",
    mean_metric_name=mean_metric_name,
    result_df=results,
    time_str=time_str,
    task=task,
    color_ind=1,
    local_base_path=local_base_path,
    plot_error_bars=False
)
```

```python
plot_two_metrics(
    first_metric_name="connectivity_ratio",
    second_metric_name=mean_metric_name,
    result_df=results,
    time_str=time_str,
    task=task,
    color_ind=1,
    plot_path=plot_path,
)
```

```python
COLORS = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897', '#f6bd60', '#e76f51', '#2a9d8f']
metric_name="connectivity_ratio"

plt.scatter(results[metric_name], results[f'{mean_metric_name}'],  c=COLORS[1], s=0.7)
plt.ylabel(f"{mean_metric_name.replace('_', ' ')}")
plt.xlabel('density')
plt.title(f"{mean_metric_name.replace('_', ' ')} of an architecture and its"
          f" density", loc='center', wrap=True)
plt.savefig(
    f"{local_base_path}/plots/{task}_teachbility_plots/{time_str}_{task}_{metric_name}_vs_{mean_metric_name}.png")
plt.show()
```
