```python
from datetime import datetime as dt
from itertools import combinations

import pandas as pd

from analsis.analsis_utils.plot_utils import (
    rads,
    greens,
    blues,
    light_oranges,
    purples,
    pinks,
    grays,
    oranges,
    plot_two_metrices_top_and_bottom_quentiles_X4,
)
```

```python
results_path = "/Volumes/noamaz/modularity/teach_archs/retina/"
local_base_path = '/Users/noamazmon/PycharmProjects/network_modularity'
results_csv_name = "2023-04-13-16-23-07_all_results_from_retina_teach_archs_results.csv"
full_results_csv_name = "2023-04-18-12-19-46_all_extended_results_from_retina_teach_archs_results.csv"
all_results = pd.read_csv(f"{results_path}/{full_results_csv_name}")
```

```python
mean_metric_name = "mean_performance"
task = "retina"
time_str = dt.now().strftime("%Y-%m-%d-%H-%M-%S")
```

```python
metircs = [
    "modularity",
    "entropy",
    "mean_num_paths_to_output_per_input_neuron",
    "mean_num_involved_neurons_in_paths_per_input_neuron",
    "mean_distances_between_input_neuron",
]
metrics_vs_metrics_combinations = list(combinations(metircs, 2))
colors = [
    rads,
    greens,
    blues,
    light_oranges,
    purples,
    pinks,
    grays,
    oranges,
]
```

```python
for couple, color in zip(metrics_vs_metrics_combinations, colors):
    plot_two_metrices_top_and_bottom_quentiles_X4(
        metric_names=couple,
        quantile_metric=mean_metric_name,
        min_connectivity_range=0.5,
        max_connectivity_range=0.8,
        all_results_df=all_results,
        time_str=time_str,
        task=task,
        color_scheme=color,
        local_base_path=local_base_path,
    )
```
